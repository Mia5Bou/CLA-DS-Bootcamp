```python
from sklearn.cluster import KMeans
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Reading the dataset
dataset = pd.read_csv('/content/drive/MyDrive/clustering_data.csv', names=['col1', 'col2'])
dataset.head()
```

```python
#Scatter plot
plt.scatter(dataset['col1'], dataset['col2']) 
```

It seems like we could have at least 3 clusters

```python
distorsions = []
for i in range(2,20):
  kmeans = KMeans(n_clusters=i).fit(dataset)
  distorsions.append(kmeans.inertia_)

#Now let's draw the elbow curve
plt.plot(range(2, 20), distorsions)
plt.grid(True)
```

According to the elbow curve, we'll have 5 clusters

```python
y_pred = KMeans(n_clusters=5, init='random', random_state=1).fit_predict(dataset)
plt.scatter(dataset['col1'], dataset['col2'], c=y_pred) 
```

And those are our five clusters ...
